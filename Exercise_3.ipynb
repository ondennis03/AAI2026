Self-Reflection Prompt for Improving Output

Tools used: Google Colab + OpenAI API + ChatGPT

```python
!pip -q install openai

import os
from google.colab import userdata
from openai import OpenAI

os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY") or ""
print("Key loaded:", bool(os.environ.get("OPENAI_API_KEY")))

client = OpenAI(api_key=os.environ.get("OPENAI_API_KEY"))

MODEL_NAME = "gpt-4.1-mini"
TEMPERATURE = 0.3

def chat(messages, model=MODEL_NAME, temperature=TEMPERATURE):
    resp = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature,
    )
    return resp.choices[0].message.content.strip()

TEXT_TO_SUMMARIZE = """
Starbucks has expanded globally while navigating challenges like shifting customer preferences,
labor issues, competition, and digital transformation. The firm emphasizes customer experience,
store operations, and partner (employee) engagement. At times, rapid growth created operational strain,
but investments in training, technology, and delivery partnerships aimed to improve convenience and loyalty.
The company also manages brand reputation and cost pressures while trying to maintain product quality.
"""

PROMPT_SUMMARY = f"""
You are a helpful assistant.

Task: Summarize the text below in 4–6 sentences.
Constraints:
- Plain language
- No bullet points
- Avoid quoting
- Keep it focused on the main ideas

TEXT:
{TEXT_TO_SUMMARIZE}
"""

initial_summary = chat([{"role": "user", "content": PROMPT_SUMMARY}])
print("=== STEP 1: INITIAL SUMMARY ===\n")
print(initial_summary)


PROMPT_CRITIQUE = f"""
You are a writing coach.

Task: Critique the SUMMARY based on the TEXT. Be specific and actionable.

Return your critique in this exact format:
SCORECARD:
- clarity (1-5):
- completeness (1-5):
- concision (1-5):
- tone (1-5):
- correctness (1-5):

WHAT'S WORKING (2 bullets max):
- ...
- ...

WHAT TO IMPROVE (3 bullets max):
- ...
- ...
- ...

TEXT:
{TEXT_TO_SUMMARIZE}

SUMMARY:
{initial_summary}
"""

critique = chat([{"role": "user", "content": PROMPT_CRITIQUE}], temperature=0.2)
print("\n=== STEP 2: SELF-REFLECTION / CRITIQUE ===\n")
print(critique)


PROMPT_IMPROVE = f"""
You are the same assistant revising your work.

Task: Rewrite the summary to address the critique.
Constraints:
- 4–6 sentences
- Plain language
- No bullet points
- No quoting
- Keep the meaning faithful to the TEXT

TEXT:
{TEXT_TO_SUMMARIZE}

CRITIQUE:
{critique}

Rewrite the improved summary only.
"""

improved_summary = chat([{"role": "user", "content": PROMPT_IMPROVE}], temperature=0.2)
print("\n=== STEP 3: IMPROVED SUMMARY ===\n")
print(improved_summary)


print("\n=== QUICK COMPARISON ===")
print("\nOriginal summary length:", len(initial_summary.split()), "words")
print("Improved summary length:", len(improved_summary.split()), "words")
```
